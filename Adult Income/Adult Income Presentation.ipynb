```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('adult.csv')
```

```python
df = df.rename(columns = {'native-country':'nativecountry'})
```

```python
highincome = df[df['income']=='>50K']
lowincome = df[df['income']=='<=50K']
```

```python
plt.hist(highincome['age'], bins=8)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Amount of high income individuals by age')
plt.ylim(0, 10000)
```

```python
plt.hist(lowincome['age'], bins=8)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Amount of low income individuals by age')
```

```python
plt.hist(highincome['educational-num'], bins=16)
plt.xlabel('Years of Education')
plt.ylabel('Count')
plt.title('Amount of high income individuals by years of education')
plt.ylim(0, 12000)
```

```python
plt.hist(lowincome['educational-num'], bins=16)
plt.xlabel('Years of Education')
plt.ylabel('Count')
plt.title('Amount of low income individuals by years of education')
```

```python
plt.hist(highincome['hours-per-week'], bins=10)
plt.xlabel('Hours worked per week')
plt.ylabel('Count')
plt.title('Amount of high income individuals by hours worked per week')
plt.ylim(0, 20000)
```

```python
plt.hist(lowincome['hours-per-week'], bins=10)
plt.xlabel('Hours worked per week')
plt.ylabel('Count')
plt.title('Amount of low income individuals by hours worked per week')
```

```python
highincome.occupation.value_counts().plot(kind = 'barh')
plt.xlabel('Count')
plt.title('Amount of high income individuals by occupation')
plt.xlim(0, 5000)
```

```python
lowincome.occupation.value_counts().plot(kind = 'barh')
plt.xlabel('Count')
plt.title('Amount of low income individuals by occupation')
```

```python
highincome.race.value_counts().plot(kind = 'barh')
plt.xlabel('Count')
plt.title('Amount of high income individuals by race')
plt.xlim(0, 30000)
```

```python
lowincome.race.value_counts().plot(kind = 'barh')
plt.xlabel('Count')
plt.title('Amount of low income individuals by race')
```

```python
highincome.gender.value_counts().plot(kind = 'barh')
plt.xlabel('Count')
plt.title('Amount of high income individuals by gender')
plt.xlim(0, 20000)
```

```python
lowincome.gender.value_counts().plot(kind = 'barh')
plt.xlabel('Count')
plt.title('Amount of low income individuals by gender')
```
